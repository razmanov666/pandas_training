# 1 Adding a new column

## 1.1 assign()

When to use assign?
- add multiple columns in a single line of code
- when you need to overwrite the values of an existing columns (best practice)

If returns a new object (a copy) with all the original columns in addition to the new ones

```python
import numpy as np
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(0, 101, size=1000)
score2 = np.random.randint(0, 101, size=1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index=np.arange(0, 1000))
series2 = pd.Series(score2, index=np.arange(0, 1000))
```

```python
# using assign() to add multiple columns
df_exams = df_exams.assign(score1=series1, score2=series2)
```

```python
df_exams
```

## 1.2 insert()

Inserts a new column at a specific position of index

```python
# using index to add a column at a specific pos
df_exams.insert(1, 'test', series1)
```
