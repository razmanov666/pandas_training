## 1. Creating a dataframe from an array

### 1.1 Option 1

```python
# creating an array
data = np.array([[1,10], [2, 15], [3, 1111]]) 
```

```python
# creating a dataframe
df = pd.DataFrame(data, 
             index=['row1', 'row2', 'row3'], 
             columns=['col1', 'col2']
            )
```

```python
# showing the dataframe
df
```

### 1.2 Option 2

```python
# creating an array with list shape
data = [[1,10], [2, 15], [3, 1111]]
```

```python
# creating a dataframe
df = pd.DataFrame(data, 
             index=['row1', 'row2', 'row3'], 
             columns=['col1', 'col2']
            )
```

```python
df
```

## 2. Creating a dataframe from a dictionary

```python
# lists used for this example
states = ['California', 'Texas', 'Florida', 'New York']
population = [123213123, 33333333, 122333333, 66664433]
```

```python
# Storing lists within a dictionary
dict_states = {'States': states, 'Population': population}
```

```python
# Creating the dataframe
df_population = pd.DataFrame(dict_states)
```

```python
# showing the dataframe
df_population
```
