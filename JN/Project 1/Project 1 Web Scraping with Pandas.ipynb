# 2 Read .csv from Multiple URLs with Pandas

https://www.football-data.co.uk/mmz4281/2425/E0.csv

https://www.football-data.co.uk/mmz4281/2425/E1.csv

https://www.football-data.co.uk/mmz4281/2425/E2.csv

Link: root + season + league

```python
# link structure
'https://www.football-data.co.uk/mmz4281' + '2425' + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = 'https://www.football-data.co.uk/mmz4281'
season = '2425'

# creating list of leagues
leagues = ["E0", "E1", "E2", "E3", "EC"]
```

```python
frames = []

# looging through leagues, read multiple csv and appedn it into a list
for league in leagues:
    df = pd.read_csv(f'{root}/{season}/{league}.csv')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd element
frames[:3]
```

```python
# create a root variable
root = 'https://www.football-data.co.uk/mmz4281'
# seasons = '2425'

# creating list of leagues
leagues = ["E0", "E1", "E2"]

frames = []

# looging through leagues, read multiple csv and appedn it into a list
for league in leagues:
    for season in range(19,25):
        df = pd.read_csv(f'{root}/{str(season) + str(season + 1)}/{league}.csv')
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 season x 3 league)
len(frames)
```

```python
# show 1st, 2nd and 3rd element (new sesason column was created)
frames[-1]
```
